```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
```

```python
train_data = datasets.MNIST('data', train="False", download="True", transform=transforms.ToTensor())
```

```python
print(train_data[1])
```

```python
for i in range(20):
    imgs = np.array(train_data[i][0], dtype='float')
    pixels = imgs.reshape((28, 28))
    plt.title(train_data[i][1])
    plt.axis('off')
    plt.imshow(pixels, cmap="gray_r")
    plt.show()
```
